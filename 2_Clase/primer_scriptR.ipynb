```r
library(repr)
library(ggplot2)
```

```r
p<-0.5
U<- runif(1,min=0,max=1)
X<- U<p
X
```

```r
sus<- 0
p<-0.5
U<- runif(1,min=0,max=1)
X<- U<p
sus <- sus+X
X+1
```

```r
n_simulations <- 100

sus<- 0
p<-0.5

fhr<-numeric(n_simulations)
for(i in 1:n_simulations){
    U<- runif(1,min=0,max=1)
    X<- U<p
    sus <- sus+X
    fhr[i]<- sus/i
}
```

```r
options(repr.plot.width=4, repr.plot.height=3)
plot(1:n_simulations,fhr)
plot(1:n_simulations,1-fhr)
```

```r
hist(fhr)
```

Creando una distribucion binomial, la idea es sumar n variables de tipo bernoulli, pues se tiene que $X=\sum_i X_i$ tiene distribución binomial si $X_i$ son de tipo bernoulli 

```r
p<-0.2
n<-5000 #numero de variables bernoulli para sumar
U<- runif(n,min=0,max=1)
X<- sum(U<p)
X
```

```r
p<-0.2
n<-50 #numero de variables bernoulli para sumar
m<-1000
fhr<-numeric(m)
for(i in 1:m){
    U<- runif(n,min=0,max=1)
    X<- sum(U<p)
    fhr[i]<-X
}
```

```r
hist(fhr,freq = F)
mean(fhr)
```

Como crear una distribucion discreta cualquiera a partir de una distribución continua



```r
x<-seq(1,10,0.1)
lambda<-1.5
plot(x,lambda*exp(-lambda*x))
plot(x,1-exp(-lambda*x))
```

```r
U<-runif(10000,min=0,max=1)
lambda<-1
X<- -log(U)/lambda
options(repr.plot.width=5, repr.plot.height=5)
hist(X,freq=F,breaks=50)
curve(dexp(x,lambda),0,7,add=T)
#lines(density(X))
mean(X)
var(X)
```
