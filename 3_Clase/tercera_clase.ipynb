```r
m <- 100
N<- 0
lambda <- 5
t <- rep(0,m)
p<- rep(0,m)
```

```r
for(i in 1:m)
{
    U=runif(1)
    x<- -1/lambda *log(U)
    N <- N+1
    t[i] <- x
    p[i] <- N
}
```

```r
plot(cumsum(t),p)
```

```r
tf<-1
m <- 100
N<- 0
lambda <- 5
t <- rep(0,m)
p<- rep(0,m)
Ns <- rep(0,1000)
tsv <- rep(0,1000)
ts <- 0
dt <- 0.1
j<- 1 
x<- 0
```

```r
while(ts<=tf)
{
    U=runif(1)
    x<- x-1/lambda *log(U)
    N <- N+1
    while(ts<=x)
        {
        Ns[j]<-N
        tsv[j]<-ts
        j<-j+1
        ts<-ts+dt
    }
    
}
```

```r
plot(tsv,Ns)
```
